```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("income.csv",names=['name','income'],skiprows=[0])
df
```

```python
df.describe()
```

```python
df.income.quantile(0.25)
df.income.quantile(0.75)
df.income.quantile(0.45)
```

```python
 df.income.quantile(1)
```

```python
percentile_99 = df.income.quantile(0.99)
percentile_99
```

```python
# df[df.income > percentile_99]
df_no_outliers = df[df.income<=percentile_99]
df_no_outliers #outlier is removed, data is clean
```
